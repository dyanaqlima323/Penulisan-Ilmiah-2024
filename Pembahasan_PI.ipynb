## Data Cleaning

```python
#Melihat nilai NaN dan Null pada data
df.isna().sum()
```

```python
# Menghapus baris 59
df = df.drop(index=59).reset_index(drop=True)
```

```python
#Mengecek ulang kolom yang sudah di-drop
df.isna().sum()
```

```python
import missingno as msno
plt.figure(figsize=(4, 4))  # Sesuaikan ukuran sesuai kebutuhan Anda
msno.matrix(df)
plt.show()
```

```python
#Melihat apakah ada duplikasi (data yang berulang)
print("Jumlah duplikasi: ", df.duplicated().sum())
```

## Exploratory Data Analysis

```python
#Melihat seluruh nama kolom
df.columns
```

```python
descriptive_stats = df[['rata rata ketinggian air (cm)', 'lama genangan', 'jumlah nilai kerugian',
                        'jiwa terdampak', 'kk terdampak', 'korban meninggal',
                        'korban hilang', 'korban luka berat', 'korban luka ringan']].describe()

print(descriptive_stats)
```

```python
# Membaca dataset
# df = pd.read_csv('path_to_your_file.csv')

# Pilih kolom numerik
numeric_columns = [
    'rata rata ketinggian air (cm)', 'lama genangan', 'rw terdampak',
    'kk terdampak', 'jiwa terdampak', 'korban meninggal', 'korban hilang',
    'korban luka berat', 'korban luka ringan', 'jumlah jiwa pengungsi tertinggi',
    'jumlah tempat pengungsian', 'jumlah nilai kerugian'
]

# Konversi kolom ke tipe data numerik jika diperlukan
for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Hitung matriks korelasi
correlation_matrix = df[numeric_columns].corr()

# Visualisasi dengan heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriks Korelasi')
plt.show()
```

```python

# Agregasi data berdasarkan tahun
yearly_impact = df.groupby('tahun')['jiwa terdampak'].sum().reset_index()

# Plot jumlah jiwa terdampak per tahun
plt.figure(figsize=(10, 6))
plt.plot(yearly_impact['tahun'], yearly_impact['jiwa terdampak'], marker='o')
plt.xlabel('Tahun')
plt.ylabel('Jiwa Terdampak')
plt.title('Jumlah Jiwa Terdampak per Tahun')
plt.grid(True)
plt.show()
```

```python
correlation = df['jumlah jiwa pengungsi tertinggi'].corr(df['jumlah tempat pengungsian'])
print("Correlation between 'jiwa' and 'jiwa':", correlation)

purple_color = '#9467bd'
plt.figure(figsize=(4, 3))
plt.scatter(df['jumlah jiwa pengungsi tertinggi'], df['jumlah tempat pengungsian'], alpha=0.5, color=purple_color)  # Menggunakan warna ungu
plt.title('Relationship between Jiwa and Tempat')
plt.xlabel('Normalized temp')
plt.ylabel('Normalized feeling temp')
plt.grid(True)
plt.show()
```
